```python
import os
```

```python
class Presentation:
    def __init__(self):
        self.slides = [
            {
                "title": "Strategic Use of Generative AI at Woolpert",
                "content": "Exploring the potential of AI to transform our business processes and client solutions"
            },
            {
                "title": "What is Generative AI?",
                "content": "• AI systems that can create new content (text, images, code, etc.)\n• Based on large language models and deep learning\n• Examples: GPT-3, DALL-E, Midjourney"
            },
            {
                "title": "Potential Applications at Woolpert",
                "content": "1. Geospatial Data Analysis\n2. Infrastructure Design Optimization\n3. Automated Report Generation\n4. Client Communication Enhancement\n5. Project Management Assistance"
            },
            {
                "title": "1. Geospatial Data Analysis",
                "content": "• Enhance satellite imagery interpretation\n• Generate 3D models from 2D data\n• Automate feature extraction and classification\n• Improve accuracy and speed of mapping processes"
            },
            {
                "title": "2. Infrastructure Design Optimization",
                "content": "• Generate multiple design options based on parameters\n• Optimize designs for efficiency and sustainability\n• Assist in creating detailed CAD models\n• Streamline the iterative design process"
            },
            {
                "title": "3. Automated Report Generation",
                "content": "• Create first drafts of technical reports\n• Summarize complex data sets\n• Generate client-ready presentations\n• Ensure consistency across documentation"
            },
            {
                "title": "4. Client Communication Enhancement",
                "content": "• Develop personalized project proposals\n• Create engaging marketing materials\n• Generate responses to RFPs more efficiently\n• Improve client interaction through AI-powered chatbots"
            },
            {
                "title": "5. Project Management Assistance",
                "content": "• Forecast project timelines and resource needs\n• Identify potential risks and suggest mitigation strategies\n• Automate routine administrative tasks\n• Enhance team collaboration through AI-powered tools"
            },
            {
                "title": "Implementation Strategy",
                "content": "1. Assess current processes and identify AI opportunities\n2. Pilot AI solutions in select projects\n3. Develop AI expertise through training and hiring\n4. Establish ethical guidelines for AI use\n5. Scale successful implementations across the organization"
            },
            {
                "title": "Challenges and Considerations",
                "content": "• Data privacy and security\n• Integration with existing systems\n• Employee training and adoption\n• Ethical use of AI in decision-making\n• Staying updated with rapidly evolving AI technologies"
            },
            {
                "title": "Next Steps",
                "content": "1. Form a cross-functional AI strategy team\n2. Conduct a detailed AI readiness assessment\n3. Identify and prioritize initial pilot projects\n4. Develop partnerships with AI solution providers\n5. Create a roadmap for long-term AI integration"
            }
        ]
        self.current_slide = 0

    def display_slide(self):
        os.system('cls' if os.name == 'nt' else 'clear')  # Clear the console
        print(f"\n{'=' * 50}")
        print(f"{self.slides[self.current_slide]['title']:^50}")
        print(f"{'=' * 50}\n")
        print(self.slides[self.current_slide]['content'])
        print(f"\n{'=' * 50}")
        print(f"Slide {self.current_slide + 1} of {len(self.slides)}")
        print("N: Next slide, P: Previous slide, Q: Quit")

    def run(self):
        while True:
            self.display_slide()
            user_input = input("Enter your choice: ").lower()
            if user_input == 'n':
                self.current_slide = (self.current_slide + 1) % len(self.slides)
            elif user_input == 'p':
                self.current_slide = (self.current_slide - 1) % len(self.slides)
            elif user_input == 'q':
                print("Thank you for attending the presentation!")
                break
            else:
                input("Invalid input. Press Enter to continue...")
```

```python
if __name__ == "__main__":
    presentation = Presentation()
    presentation.run()
```
